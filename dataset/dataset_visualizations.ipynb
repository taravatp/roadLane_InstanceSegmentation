```
import os.path as ops
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
gt_original = '/content/drive/MyDrive/Lane_Detection/TUSMIPLE/training/gt_image/0013.png'
gt_binary = '/content/drive/MyDrive/Lane_Detection/TUSMIPLE/training/gt_binary_image/0013.png'
gt_instance = '/content/drive/MyDrive/Lane_Detection/TUSMIPLE/training/gt_instance_image/0013.png'

gt_original = cv2.imread(gt_original, cv2.IMREAD_UNCHANGED)
gt_original = cv2.cvtColor(gt_original, cv2.COLOR_RGB2BGR)

gt_binary = cv2.imread(gt_binary, cv2.IMREAD_UNCHANGED)
gt_instance = cv2.imread(gt_instance, cv2.IMREAD_UNCHANGED)
```

```
plt.figure(figsize=(30,10))
ax1 = plt.subplot(131)
plt.imshow(gt_original)
plt.title('original image')
ax1 = plt.subplot(132)
plt.imshow(gt_binary, cmap='gray')
plt.title('ground truth binary label')
ax1 = plt.subplot(133)
plt.imshow(gt_instance,cmap='gray')
plt.title('ground truth instance label')
```
